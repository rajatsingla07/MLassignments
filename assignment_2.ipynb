```python
import numpy as np
#Q1 to make array of (10,1) using numpy

a=np.random.rand(10,1)
print(a)
print('--'*50)


```

```python
#Q2 To make array of (20,1) and find variance and standard deviation
a=np.random.random((20,1))
print(a)
print('--'*50)
#standard deviation
print(np.std(a))
print('--'*50)
#variance
print(np.std(a)**2)


```

```python
#Q3 Matrix multiplication and sum
a=np.random.rand(10,20)
b=np.random.rand(20,25)
m=np.matmul(a,b)
print('--'*70)
print(m)
print('--'*70)
print(m.shape)
print('--'*70)
print(np.sum(m))

```

```python

#Q4 to make an array and a function as (1/1+exp(-x))
a= np.random.rand(10,1)
print(a)
print('--'*50)
def f(x):
    return (1 / (1 + np.exp(-x)))

r = np.apply_along_axis(f, 0, a)
print(r)
```
